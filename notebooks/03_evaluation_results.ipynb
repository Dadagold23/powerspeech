## 4. Visualise WER across samples

```python
import matplotlib.pyplot as plt

samples  = [d['audio_file'] for d in results['detailed']]
wer_vals = [d['transcription'].get('WER', 0) * 100 for d in results['detailed']]

plt.figure(figsize=(8, 4))
plt.bar(samples, wer_vals, color='steelblue')
plt.axhline(results['avg_WER'] * 100, color='red', linestyle='--', label='Average WER')
plt.ylabel('WER (%)')
plt.title('Word Error Rate by Sample')
plt.legend()
plt.tight_layout()
plt.savefig('../docs/wer_chart.png', dpi=150)
plt.show()
```
