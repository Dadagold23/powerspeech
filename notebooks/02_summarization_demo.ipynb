## 1. Instantiate and load the pipeline

```python
summarizer = SummarizationModule()
print('Model:', summarizer.model_name)
```

## 2. Summarize a sample transcript

Paste your transcript text below, or load it from a saved session file.

```python
TEXT = """
Artificial intelligence is rapidly transforming industries around the world.
In healthcare, AI models can detect diseases from medical images with accuracy
that rivals experienced clinicians. In finance, algorithmic trading systems
process millions of transactions per second. Natural language processing
enables machines to understand and generate human language, powering virtual
assistants, translation services, and content generation tools.
Despite these advances, challenges remain around bias, explainability, and
the ethical use of AI systems in high-stakes decision-making.
"""

summary = summarizer.summarize(TEXT)
print('Summary:\n', summary)
```

## 3. Extract key bullet points (extractive)

```python
points = summarizer.get_key_points(TEXT, num_points=4)
for i, p in enumerate(points, 1):
    print(f'{i}. {p}')
```

## 4. Text chunking behaviour for long inputs

```python
long_text = TEXT.strip() * 10  # simulate a long transcript
chunks = SummarizationModule._chunk_text(long_text, max_chunk_chars=3000)
print(f'Long text ({len(long_text)} chars) split into {len(chunks)} chunk(s)')
```
