# Speech-to-Text Module Demo

Demonstrates the `SpeechToTextModule` using OpenAI Whisper.

```python
import sys
sys.path.insert(0, '..')
from src.speech_to_text import SpeechToTextModule
```

## 1. Instantiate the module

The model is loaded lazily on first use.

```python
stt = SpeechToTextModule(model_name='base')
print('Module ready. Model:', stt.model_name)
```

## 2. Transcribe an audio file

Update `AUDIO_PATH` to point to one of your `.wav` files in `data/samples/`.

```python
AUDIO_PATH = '../data/samples/sample_01.wav'  # replace with your file

result = stt.transcribe(AUDIO_PATH)
print('Language detected:', result['language'])
print('\nTranscript:\n', result['text'])
```

## 3. Inspect timestamped segments

```python
for seg in result['segments'][:5]:
    print(f"[{seg['start']:.1f}s -> {seg['end']:.1f}s]  {seg['text'].strip()}")
```

## 4. List supported languages

```python
langs = stt.get_supported_languages()
print(f'Whisper supports {len(langs)} languages:')
print(', '.join(sorted(langs)[:20]), '...')
```
